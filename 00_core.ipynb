# card -- Basic Playing Card

> A simple API for creating and using playing cards

```python
#|hide
from nbdev.showdoc import *
```

```python
#|export
import attrs
from dataclasses import dataclass
def foo(): pass
```

```python
#| export
suits = ["♣️","♦️","♥️","♠️"]
ranks = [None, "A"] + [str(x) for x in range(2,11)] + ["J", "Q", "K"]
```

```python
#| export
class Card:
    "Playing Card"
    def __init__(self, suit, rank):
        self.suit, self.rank = suit, rank
    def __str__(self): return f"{ranks[self.rank]}{suits[self.suit]}"
    __repr__ = __str__
```

```python
#| export
@attrs.define
class CardAttrs:
    "Playing Card"
    suit: str
    rank: str
```

```python
#| export
@dataclass
class CardDataclass:
    "Playing Card"
    suit: str
    rank: str
```

```python
a = Card(suit=1, rank=3)
a
```

```python
b = CardAttrs(suit=1, rank=3)
b
```

```python
c = CardDataclass(suit=1, rank=3)
c
```
